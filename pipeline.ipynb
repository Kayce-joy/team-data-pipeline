```python
import pandas as pd
import numpy as np

# Sample data for Student A's step
data = {
    'Name': ['john', 'jane', 'jim', 'jack', 'jill'],
    'age': [20, 21, 22, 23, 24],
    'weight': [150, 155, 160, 165, 170],
}

df = pd.DataFrame(data)
df
```

```python
#Summary statistics for numeric columns
stats = df.describe()

# Calculate average (mean) values for age and weight
averages = df.mean(numeric_only=True)

# Calculate median values
median_age = df['age'].median()
median_weight = df['weight'].median()

# Calculate mode values
mode_age = df['age'].mode()[0]
mode_weight = df['weight'].mode()[0]

print("Descriptive Statistics:\n", stats)
print("\nAverage Values:\n", averages)
print("\nMedian Values:\n Age:", median_age, " Weight:", median_weight)
print("\nMode Values:\n Age:", mode_age, " Weight:", mode_weight)
```
